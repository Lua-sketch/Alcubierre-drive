# The Alcubierre Drive

Here we show how one might visualize the geometry of the warp drive solution proposed by Alcubierre.

First, we import the libraries we will need. For the math functions (such as square root and hyperbolic trigonometric functions), we will need numpy. For plotting, we'll use matplotlib.pyplot. Finally, because we are making a 3D graph, we'll also import some 3D tools from matplotlib.

```python
from mpl_toolkits import mplot3d #these are the 3d tools mentioned above
import numpy as np #we can now reference numpy functions as np
import matplotlib.pyplot as plt #we can now reference matplotlib.pyplot as plt
```

Now, we need to cover some of the math behind the warp drive. First, note that the metric for his geometry (in Cartesian coordinates and natural units, G=c=1), is given by

ds²= -dt²+dx²+dy²+(dz-v(t)f(x, y, z-z₀(t))dt)²

where f is a form function with value 0 outside the bubble and value 1 inside of it. Alcubierre chose spherically symmetric form functions, f(r(t)), r(t)=((z-z₀(t))²+x²+y²)¹/². 

We consider the case of

f(r)=(tanh(σ(r+R))-tanh(σ(r-R)))/(2tanh(σR))

where R is the radius of the warp bubble and σ is inversely proportional to the thickness of the bubble wall. If σ is sufficiently large, f(r)=1 if r ∈ [0, R] and f(r)=0 if r ∈ (R, ∞).

The expansion of the volume elements is given by

θ = v((z-z₀)/r)(df(r)/dr).

In order to plot this function, we need to choose a fixed value for z₀, v, R, and σ. Let's plot the particular case in which v=R=1, z₀=0, and σ=8. Then,

θ = (4/tanh(8))(z/r)(sec²(8(r+1))-sec²(8(r-1))) = (4/tanh(8))(z/(√ρ²+z²))(sec²(8((√ρ²+z²)+1))-sec²(8((√ρ²+z²)-1)))

Let's begin our program by defining this volume expansion element as a function of z and ρ. The reason is that these will be two of our axes.

```python
def theta(z, rho):
    return (4/np.tanh(8))*(z/np.sqrt(z**2+rho**2))*((1/np.cosh(8*(np.sqrt(z**2+rho**2)+1)))**2-(1/np.cosh(8*(np.sqrt(z**2+rho**2)-1)))**2)
```

Now, we can use numpy to create the values for z and ρ that we wish to plot.

```python
z = np.linspace(-4, 4, 160) #gives a list of 160 equally spaced values between -4 and 4
rho = np.linspace(-4, 4, 160) #does the same

Z, RHO = np.meshgrid(z, rho) #creates two matrices from the two lists above
```

Next, compute the values of theta that we will plot:

```python
THETA=theta(Z, RHO)
```

Finally, plot the figure:

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot_wireframe(Z, RHO, THETA, color ='black')
#ax.set_title('Volume plot for the Alcubierre drive')
```

The image looks tiny... let's try to change the axes:

```python
z2 = np.linspace(-3, 3, 160) #gives a list of 160 equally spaced values between -4 and 4
rho2 = np.linspace(-3, 3, 160) #does the same

Z2, RHO2 = np.meshgrid(z2, rho2) #creates two matrices from the two lists above

THETA2=theta(Z2, RHO2)
fig2 = plt.figure()
ax2 = plt.axes(projection ='3d')
ax2.set_xlim([-3, 3])
ax2.set_ylim([-3, 3])
ax2.plot_wireframe(Z2, RHO2, THETA2, color ='black')
#ax2.set_title('Volume plot for the Alcubierre drive')
```
